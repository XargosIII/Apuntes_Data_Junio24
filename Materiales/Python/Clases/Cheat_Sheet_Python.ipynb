```python
# Bucle `for`
# Iterar sobre una lista
for elemento in [1, 2, 3, 4]:
    print(elemento)  # Salida: 1 2 3 4

# Iterar sobre un diccionario
mi_diccionario = {'a': 1, 'b': 2}
for clave, valor in mi_diccionario.items():
    print(clave, valor)  # Salida: a 1  b 2

# Bucle `while`
# Contador básico
contador = 0
while contador < 5:
    print(contador)  # Salida: 0 1 2 3 4
    contador += 1

# Condicionales
# Evaluación de una variable
x = 10
if x > 0:
    print("x es positivo")  # Salida: x es positivo
elif x == 0:
    print("x es cero")
else:
    print("x es negativo")

# Listas Comprimidas
# Crear una lista de cuadrados
cuadrados = [x**2 for x in range(10)]
print(cuadrados)  # Salida: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# Filtrar elementos pares
pares = [x for x in range(10) if x % 2 == 0]
print(pares)  # Salida: [0, 2, 4, 6, 8]

# Tuplas
# Definir y acceder a una tupla
mi_tupla = (1, 2, 3)
print(mi_tupla[0])  # Salida: 1

# Conjuntos
# Operaciones con conjuntos
mi_conjunto = {1, 2, 3, 4}
mi_conjunto.add(5)
mi_conjunto.remove(3)
print(mi_conjunto)  # Salida: {1, 2, 4, 5}

# Diccionarios
# Operaciones con diccionarios
mi_diccionario = {'clave1': 'valor1', 'clave2': 'valor2'}
print(mi_diccionario['clave1'])  # Salida: valor1

mi_diccionario['clave3'] = 'valor3'
print(mi_diccionario)  # Salida: {'clave1': 'valor1', 'clave2': 'valor2', 'clave3': 'valor3'}

# Funciones Básico
# Definir y llamar a una función
def mi_funcion(parametro1, parametro2):
    return parametro1 + parametro2

resultado = mi_funcion(5, 3)
print(resultado)  # Salida: 8

# Función con valor por defecto
def saludar(nombre="Mundo"):
    print(f"Hola, {nombre}!")

saludar()  # Salida: Hola, Mundo!
saludar("Python")  # Salida: Hola, Python!


# Función lambda (anónima)
suma = lambda a, b: a + b
print(suma(2, 3))  # Salida: 5


```
