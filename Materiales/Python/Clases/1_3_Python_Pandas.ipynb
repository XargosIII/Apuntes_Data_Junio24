#### **`map()`**
La funci칩n **`map()`** nos va a facilitar hacer bucles constantemente. Permite aplicar una **funci칩n** a cada uno de los elementos de una lista, o cualquier otra estructura que sea iterable (lista, set, tupla...). Ahora que estamos conociendo la librer칤a pandas y vamos a estar manej치ndonos con **`Series`** y con **`DataFrame`** vamos a ver su gran utilidad.

Hay que tener en cuenta que una `Series` es una columna y su comportamiento es similar al de una lista. **`map()`** va a facilitarnos el trabajo de transformaci칩n de datos, si hay que hacer una modificaci칩n para toda una columna (o varias) a trav칠s de una funci칩n.

**Ejemplo**
```python
# Definimos una funci칩n que calcula el cuadrado de un n칰mero
def cuadrado(numero):
    return numero ** 2

# Tenemos una lista de n칰meros
numeros = [1, 2, 3, 4, 5]

# Usamos map() para aplicar la funci칩n cuadrado a cada elemento de numeros
# casteamos para que el resultado sea una lista
resultados = list(map(cuadrado, numeros))

print(resultados)
```
La salida de este print ser치 `[1, 4, 9, 16, 25]`. Sin hacer un bucle se ha llamado a la funci칩n aplic치ndola para cada uno de los elementos que hay en la lista.

`````python
nuevalista = list(map(euros,lista)) #map() puede evitarnos bucles tediosos
nuevalista
`````

**`map()`** puede facilitarnos mucho la vida para hacer transformaciones, o preparaciones de datos. Es una de las alternativas eficientes que hay para hacer estas modificaciones. En la siguiente celda de c칩digo vamos a ver un ejemplo para convertir una variable categ칩rica, con nuestros amigos aguacates, que pase de string a int. En la columna `type` hay informaci칩n sobre dos posibles tipos de aguacates: `'conventional'` y `'organic'`. Puede ser 칰til tener esta variable categ칩rica en formato num칠rico. Por no perder informaci칩n lo que vamos a hacer es:
1. Crear un diccionario que sirva para mapear. El funcionamiento es el siguiente:
  * el diccionario es par de valores `key:value`
  * el `key` es el valor que encuentra (`'conventional'` u `'organic'`)
  * el `value` es el valor que va a poner en caso de encontrar uno u otro (1 칩 2)
1. A침adir una columna `type_numeric` en la que tengamos el tipo de aguacate en formato num칠rico. En esta columna debe haber informaci칩n num칠rica, para `'conventional'` que sea **1** y para `'organic'` que sea **2**.

`````python
# confirmaci칩n que en type hay 칰nicamente esos dos valores
df.type.unique()
`````

`````python
# diccionario para los nuevos valores
dic_mapeo = {
            'conventional': 1,
            'organic': 2
            }
`````

`````python
# creaci칩n nueva columna
# crea la columna 'type_numeric'
# con la informaci칩n de la columna 'type'
# donde encuentres la key, pon el value de lo que hay en el diccionario
df['type_numeric'] = df['type'].map(dic_mapeo)
`````

##GroupBy (nos suena?? 游뱂)
GroupBy de Pandas es una funci칩n potente y vers치til en Python. Nos permite dividir sus datos en grupos separados para realizar c치lculos para un mejor an치lisis.
Un DataFrame puede agruparse en sus filas (eje=0) o en sus columnas (eje=1). Una vez hecho esto, se aplica una funci칩n a cada grupo, produciendo un nuevo valor. Finalmente, los resultados de todas esas aplicaciones de funciones se combinan en un objeto resultado. La forma del objeto resultante depender치 normalmente de lo que se haga con los datos.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html

Vamos a crear un df sencillito para entender bien como va la cosa

`````python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'], 'key2' : ['one', 'two', 'one', 'two', 'one'],'data1' : np.random.randn(5),'data2' : np.random.randn(5)})
`````

`````python
df.head()
`````

Los datos (una Serie) han sido agregados seg칰n la clave de grupo, produciendo una nueva Serie que ahora est치 indexada por los valores 칰nicos de la columna clave1. El 칤ndice resultante tiene el nombre 'clave1' porque la columna DataFrame `df['clave1']` lo hizo.
